```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
# the func returns the cardinality of the intersection of a and b
start_time=timeit.default_timer()
print(func4([1,2,3,4],[2,4,6,8]))
print('time elapsed:',(timeit.default_timer()-start_time)*1000,'ms')
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
# the function returns the square root of the given integer number
start_time=timeit.default_timer()
print(func5(25))
print('time elapsed:',(timeit.default_timer()-start_time)*1000,'ms')
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)
    return A
def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - i + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
    elif n > 2:
        m = n // 3
        func8_helper(A, i, j - m)
        func8_helper(A, i + m, j)
        func8_helper(A, i, j - m)
    return A
```

```python
# the function sorts the given list of numbers
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# the function shifts the elements of the array k times in cycling order
```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
start_time=timeit.default_timer()
print(func12([1,2,3,4,5],2))
print('time elapsed:',(timeit.default_timer()-start_time)*1000,'ms')
```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  result=0
  for i in range(0,len(arr)):
    result+=arr[i]
  return result
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max=arr[0]
  min=arr[0]
  for i in range(0,len(arr)):
    if arr[i]>max:
      max=arr[i]
    if arr[i]<min:
      min=arr[i]
  return max,min
```

```python
# call sum_arr to find sum of elements in given array for arr = [12, 1234, 45, 67, 1]
arr = [12,1234,45,67,1]
print(max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_evenOdd(arr):
  count_even=0
  count_odd=0
  for i in range(0,len(arr)):
    if arr[i]%2==0:
      count_even+=1
    else:
      count_odd+=1
  return count_even,count_odd
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,key):
  for i in range(0,len(arr)):
    if arr[i]==key:
      return 1
  return -1
```
